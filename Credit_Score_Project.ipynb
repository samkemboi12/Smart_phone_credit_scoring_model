### 🆔 Identity / Structure Variables

These columns identify the loan and define the time context of each record.

*   **LOAN_ID** → Unique identifier for each loan (key column)
*   **DATE** → Reporting (snapshot) date on which all loan metrics are calculated
*   **SALE_DATE** → Loan origination date (when the device was sold)

### 💰 Payment Behavior (Core Risk Signals)

These variables describe how customers are repaying their loans over time.

*   **TOTAL_PAID** → Cumulative amount paid by the customer up to the snapshot date
*   **TOTAL_DUE_TODAY** → Total amount the customer is expected to have paid by this date
*   **DEPOSIT** → at loan start (structural / contract feature)
*   **ADVANCE** → Amount the customer has paid ahead of schedule (early payments after loan start)  → during loan lifecycle (behavioral / repayment feature)

### 📝 Contract / Pricing Variables

These describe the structure of the loan agreement.

*   **WEEKLY_RATE** → Scheduled weekly repayment amount for the loan
*   **INITIAL_PAY** → First payment made at the start of the loan
*   **EXPECTED_PAYMENT** → Payment amount expected for the reporting period
*   **DISCOUNT** → Discount applied to the loan or cash price
*   **CREDIT_CHECK_DONE** → Result or status of the credit check performed

### Columns to DROP (as they reveal the result or are problematic)

These columns are identified as potentially problematic, either because they directly reveal information that a predictive model should ideally infer (data leakage), or because they have too many missing values to be useful.

*   **CLOSING_BALANCE**
    *   **Reason:** Shows the final loan state, which can lead to data leakage.
*   **BALANCE**
    *   **Reason:** Can indirectly reveal whether the customer paid well or not, making it difficult to predict for new customers.
*   **BALANCE_DUE_STATUS**
    *   **Reason:** Already indicates if a customer is behind on payments.
*   **DAYS_PAST_DUE**
    *   **Reason:** Directly shows how late the customer is.
    *   **Decision Rationale:** Both `BALANCE_DUE_STATUS` and `DAYS_PAST_DUE` inherently contain the outcome we might want to predict (customer risk), making them unsuitable as features for a forward-looking model.
*   **ARREARS**
    *   **Reason:** Directly shows the unpaid amount, similar to the above points.
*   **PAYMENT_AMOUNT**
    *   **Reason:** Almost all values are missing (~99.8% missing, 30 non-null out of 20,742), providing no signal and just noise.
*   **ADJUSTMENT_AMOUNT**
    *   **Reason:** Almost all values are missing, similar to `PAYMENT_AMOUNT`.
*   **PREPAYMENT_AMOUNT**
    *   **Reason:** While it represents early/upfront payment, `INITIAL_PAY` (money paid at the beginning of the loan) is already kept. Keeping both is redundant for upfront payments.
*   **OVERPAYMENT_AMOUNT**
    *   **Reason:** About 97% of rows are zero, creating highly unbalanced/skewed data. It should be used only as a small secondary signal, not as a major feature.
*   **EXPECTED_PAYMENT**
    *   **Reason:** Many rows are missing or zero (~88%). However, when values exist, they contain useful information. Should be treated carefully if kept.
*   **BALANCE_DUE_TO_DATE**
    *   **Reason:** Directly indicates if the customer is ahead or behind on payments. It is a strong behavioral/performance signal that is very close to the final outcome (good or bad customer).
*   **CREDIT_EXPIRY**
    *   **Reason:** Does not directly tell how the customer behaves, pays, or if they are risky. It is a contractual date, not a behavioral indicator.
*   **TOTAL_PAID_WITH_ADJUSTMENTS_15D**
    *   **Reason:** If the goal is to predict if a customer will become risky, this column already reveals how well they performed in the early critical window, leading to data leakage. The model would learn the outcome directly from this feature.
*   **RETURN_DATE**
    *   **Reason:** Too many missing values (91%).

### FEATURE ENGINEERING

#### 1. Target Variables (`engineer_target_variables`)
These features are different definitions of 'default' or payment performance, allowing for flexibility in modeling objectives.

*   `TARGET_DEFAULT_BINARY`: A binary flag (0 or 1) indicating if the account status is 'PAR 7', 'PAR 30', or 'Inactive' based on `ACCOUNT_STATUS_L2`.
*   `TARGET_SEVERE_DEFAULT`: A binary flag (0 or 1) indicating if the account status is 'PAR 30' (severe delinquency) based on `ACCOUNT_STATUS_L2`.
*   `TARGET_FPD`: A binary flag (0 or 1) indicating if the loan has a 'First Payment Default' based on `ACCOUNT_STATUS_L1`.
*   `TARGET_DPD`: The number of `DAYS_PAST_DUE`, representing a continuous target for regression tasks (filled with 0 for non-missing values).

```
# TARGET VARIABLE ENGINEERING

def engineer_target_variables(credit_df):
    """
    Create multiple target definitions for default prediction.
    Returns dataframe with target columns.
    """
    df = credit_df.copy()

    # Target 1: Binary default based on ACCOUNT_STATUS_L2
    # PAR = Past Due / Problematic accounts
    df['TARGET_DEFAULT_BINARY'] = df['ACCOUNT_STATUS_L2'].isin(['PAR 7', 'PAR 30', 'Inactive']).astype(int)

    # Target 2: Severe default (PAR 30+ - serious delinquency)
    df['TARGET_SEVERE_DEFAULT'] = df['ACCOUNT_STATUS_L2'].isin(['PAR 30']).astype(int)

    # Target 3: Early-stage delinquency (FPD - First Payment Default)
    df['TARGET_FPD'] = df['ACCOUNT_STATUS_L1'].str.contains('FPD', na=False).astype(int)

    # Target 4: Days past due as continuous target (regression)
    df['TARGET_DPD'] = df['DAYS_PAST_DUE'].fillna(0)

    return df

credit = engineer_target_variables(clean_credit_snapshot)

```

#### 2. Payment Behavior Features (`engineer_payment_behavior_features`)
These features capture aspects of how customers make payments, including ratios, gaps, and consistency.

*   `PAYMENT_RATIO`: The ratio of `TOTAL_PAID` to `TOTAL_DUE_TODAY`, indicating payment fulfillment.
*   `PAYMENT_GAP`: The difference between `TOTAL_DUE_TODAY` and `TOTAL_PAID`, showing how much is unpaid or overpaid.
*   `PAYMENT_ADVANCE_RATIO`: The ratio of `ADVANCE` amount to `TOTAL_DUE_TODAY`, reflecting proactive payments.
*   `ARREARS_TO_DUE_RATIO`: The ratio of `ARREARS` to `TOTAL_DUE_TODAY`, indicating the severity of outstanding arrears.
*   `ARREARS_TO_BALANCE_RATIO`: The ratio of `ARREARS` to `BALANCE`, showing arrears relative to the current outstanding balance.
*   `PAYMENT_MADE`: A binary flag (0 or 1) indicating if a `PAYMENT` was made on the snapshot date.
*   `CUMULATIVE_PAYMENTS`: The cumulative count of payments made over the loan's history up to the snapshot date.
*   `TOTAL_EXPECTED_PAYMENTS`: The cumulative sum of `EXPECTED_PAYMENT` over the loan's history.
*   `PAYMENT_CONSISTENCY_SCORE`: A score reflecting the consistency of payments, calculated as `CUMULATIVE_PAYMENTS` divided by the number of snapshots for the loan.
*   `ARREARS_CHANGE`: The daily change in `ARREARS`, indicating whether arrears are increasing or decreasing.
*   `ARREARS_INCREASING`: A binary flag (0 or 1) if `ARREARS_CHANGE` is positive, indicating worsening arrears.


```
# ============================================
# FEATURE ENGINEERING
# ============================================

def engineer_payment_behavior_features(credit_df):
    """Features capturing customer payment behavior."""
    df = credit_df.copy()

    # 1. Payment ratio metrics
    df['PAYMENT_RATIO'] = df['TOTAL_PAID'] / (df['TOTAL_DUE_TODAY'] + 1)
    df['PAYMENT_GAP'] = df['TOTAL_DUE_TODAY'] - df['TOTAL_PAID']
    df['PAYMENT_ADVANCE_RATIO'] = df['ADVANCE'] / (df['TOTAL_DUE_TODAY'] + 1)

    # 2. Arrears severity
    df['ARREARS_TO_DUE_RATIO'] = df['ARREARS'] / (df['TOTAL_DUE_TODAY'] + 1)
    df['ARREARS_TO_BALANCE_RATIO'] = df['ARREARS'] / (df['BALANCE'] + 1)

    # 3. Payment consistency (rolling metrics per loan)
    df = df.sort_values(['LOAN_ID', 'DATE'])
    df['PAYMENT_MADE'] = (df['PAYMENT'] > 0).astype(int)

    # Cumulative payment consistency
    df['CUMULATIVE_PAYMENTS'] = df.groupby('LOAN_ID')['PAYMENT_MADE'].cumsum()
    df['TOTAL_EXPECTED_PAYMENTS'] = df.groupby('LOAN_ID')['EXPECTED_PAYMENT'].cumsum()
    df['PAYMENT_CONSISTENCY_SCORE'] = df['CUMULATIVE_PAYMENTS'] / (df.groupby('LOAN_ID').cumcount() + 1)


    # 5. Rolling arrears trend (increasing or decreasing)
    df['ARREARS_CHANGE'] = df.groupby('LOAN_ID')['ARREARS'].diff()
    df['ARREARS_INCREASING'] = (df['ARREARS_CHANGE'] > 0).astype(int)

    return df
```

#### 3. Temporal Features (`engineer_temporal_features`)
These features derive information from dates, capturing loan age, remaining term, and seasonality.


*   `DAYS_TO_EXPIRY`: The number of days remaining until `CREDIT_EXPIRY`.
*   `DAYS_PAST_NEXT_INVOICE`: The number of days `DATE` is past `NEXT_INVOICE_DATE`.
*   `IS_PAYMENT_LATE`: Binary flag (0 or 1) if `DAYS_PAST_NEXT_INVOICE` is greater than 0.


```
def engineer_temporal_features(credit_df):
    """Features based on time and loan age."""
    df = credit_df.copy()

    # 1. Remaining term
    df['DAYS_TO_EXPIRY'] = (df['CREDIT_EXPIRY'] - df['DATE']).dt.days

    # 2. Payment timing relative to due date
    df['DAYS_PAST_NEXT_INVOICE'] = (df['DATE'] - df['NEXT_INVOICE_DATE']).dt.days
    df['IS_PAYMENT_LATE'] = (df['DAYS_PAST_NEXT_INVOICE'] > 0).astype(int)

    return df
```

#### 4. Financial Ratios (`engineer_financial_ratios`)
These features focus on financial health, leverage, and specific payment behaviors related to initial and regular payments.

*   `INITIAL_PAY_RATIO`: Ratio of `INITIAL_PAY` to `WEEKLY_RATE`, showing initial commitment relative to regular payments.
*   `DEPOSIT_RATIO`: Ratio of `DEPOSIT` to `WEEKLY_RATE`.
*   `TOTAL_OBLIGATION`: An approximation of the annual obligation based on `WEEKLY_RATE`.
*   `HAS_OVERPAYMENT`: Binary flag (0 or 1) if `OVERPAYMENT_AMOUNT` is greater than 0.
*   `BALANCE_TO_OBLIGATION`: Ratio of `BALANCE` to `WEEKLY_RATE`, indicating outstanding balance relative to weekly payment.
*   `BALANCE_REDUCTION_RATE`: The rate at which the `BALANCE` is being reduced over time.
*   `NET_PAYMENT_POSITION`: The difference between `ADVANCE` and `ARREARS`, indicating net financial standing.
*   `POSITION_STATUS`: Categorical feature ('Ahead', 'Behind', 'OnTrack') based on `NET_PAYMENT_POSITION`.

```
def engineer_financial_ratios(credit_df):
    """Features capturing financial health and leverage."""
    df = credit_df.copy()

    # 1. Initial payment effectiveness
    df['INITIAL_PAY_RATIO'] = df['INITIAL_PAY'] / (df['WEEKLY_RATE'] + 1)
    df['DEPOSIT_RATIO'] = df['DEPOSIT'] / (df['WEEKLY_RATE'] + 1)

    # 2. Loan structure metrics
    df['TOTAL_OBLIGATION'] = df['WEEKLY_RATE'] * 52  # Approximate annual obligation

    # 3. Overpayment patterns
    df['HAS_OVERPAYMENT'] = (df['OVERPAYMENT_AMOUNT'] > 0).astype(int)

    # 4. Balance management
    df['BALANCE_TO_OBLIGATION'] = df['BALANCE'] / (df['WEEKLY_RATE'] + 1)
    df['BALANCE_REDUCTION_RATE'] = -df.groupby('LOAN_ID')['BALANCE'].diff() / (df['BALANCE'] + 1)

    # 5. Advanced vs Arrears net position
    df['NET_PAYMENT_POSITION'] = df['ADVANCE'] - df['ARREARS']
    df['POSITION_STATUS'] = np.where(df['NET_PAYMENT_POSITION'] > 0, 'Ahead',
                                     np.where(df['NET_PAYMENT_POSITION'] < 0, 'Behind', 'OnTrack'))

    return df
```

#### 5. Risk Scores (`engineer_risk_scores`)
These are composite scores designed to aggregate multiple signals into a single risk indicator.

*   `PAYMENT_BEHAVIOR_SCORE`: A composite score (0-100) combining `PAYMENT_RATIO`, `ARREARS_TO_DUE_RATIO`, `ADVANCE` flag, and `PAYMENT_CONSISTENCY_SCORE`.
*   `EARLY_WARNING_SCORE`: A score (0-100) combining `DAYS_PAST_DUE` flag, `ARREARS_INCREASING` flag, and `PAYMENT_RATIO` below a threshold, designed to detect early signs of trouble.
*   `CUSTOMER_STABILITY_SCORE`: A score (0-100) combining `CREDIT_CHECK_DONE` status, `PAYMENT_CONSISTENCY_SCORE`, and `CUSTOMER_AGE`.
*   `RISK_TIER`: Categorical feature ('High Risk', 'Medium Risk', 'Low Risk', 'Minimal Risk') based on `EARLY_WARNING_SCORE`.


```
def engineer_risk_scores(credit_df):
    """Composite risk scores from multiple signals."""
    df = credit_df.copy()

    # 1. Payment behavior score (0-100, higher = better)
    df['PAYMENT_BEHAVIOR_SCORE'] = np.clip(
        100 * (
            (df['PAYMENT_RATIO'] * 0.4) +
            (1 - np.minimum(1, df['ARREARS_TO_DUE_RATIO'])) * 0.3 +
            (df['ADVANCE'] > 0).astype(int) * 0.2 +
            df['PAYMENT_CONSISTENCY_SCORE'] * 0.1
        ), 0, 100
    )

    # 2. Early warning score (detects potential default before it happens)
    df['EARLY_WARNING_SCORE'] = np.clip(
        (
            (df['DAYS_PAST_DUE'] > 0).astype(int) * 30 +
            (df['ARREARS_INCREASING'] == 1).astype(int) * 25 +
            (df['PAYMENT_RATIO'] < 0.8).astype(int) * 20
        ), 0, 100
    )

    # 3. Customer stability score
    df['CUSTOMER_STABILITY_SCORE'] = np.clip(
        (
            (df['CREDIT_CHECK_DONE'] == 'Pass').astype(int) * 40 +
            (df['PAYMENT_CONSISTENCY_SCORE'] * 30) +
            ((df['CUSTOMER_AGE'] / 365) * 30)
        ), 0, 100
    )

    # 4. Delinquency risk tier
    df['RISK_TIER'] = np.where(
        df['EARLY_WARNING_SCORE'] >= 70, 'High Risk',
        np.where(df['EARLY_WARNING_SCORE'] >= 40, 'Medium Risk',
                 np.where(df['EARLY_WARNING_SCORE'] >= 20, 'Low Risk', 'Minimal Risk'))
    )

    return df
```

#### 6. First Payment Features (`engineer_first_payment_features`)
These features are specifically designed to capture behavior around the initial stages of the loan, particularly for First Payment Default (FPD).

*   `FIRST_PAYMENT_RATIO`: Ratio of `FIRST_PAYMENT` to `FIRST_EXPECTED_PAYMENT`.
*   `FIRST_PAYMENT_SHORTFALL`: The difference between `FIRST_EXPECTED_PAYMENT` and `FIRST_PAYMENT` if `FIRST_EXPECTED_PAYMENT` is greater.
*   `EARLY_15D_PAYMENT_RATIO`: Ratio of `TOTAL_PAID_WITH_ADJUSTMENTS_15D` to an approximation of two weeks of `WEEKLY_RATE`.
*   `EARLY_15D_ADEQUATE`: Binary flag (0 or 1) if `EARLY_15D_PAYMENT_RATIO` is 0.5 or more.
*   `INITIAL_DEPOSIT_GAP`: The difference between `INITIAL_PAY` and `DEPOSIT`.
*   `INITIAL_DEPOSIT_RATIO`: Ratio of `INITIAL_PAY` to `DEPOSIT`.

```
def engineer_first_payment_features(credit_df):
    """First Payment Default (FPD) specific features."""
    df = credit_df.copy()

    # 1. First payment performance
    df['FIRST_PAYMENT_RATIO'] = df['FIRST_PAYMENT'] / (df['FIRST_EXPECTED_PAYMENT'] + 1)
    df['FIRST_PAYMENT_SHORTFALL'] = np.maximum(0, df['FIRST_EXPECTED_PAYMENT'] - df['FIRST_PAYMENT'])

    # 2. Early payment behavior (first 15 days)
    df['EARLY_15D_PAYMENT_RATIO'] = df['TOTAL_PAID_WITH_ADJUSTMENTS_15D'] / (df['WEEKLY_RATE'] * 2 + 1)
    df['EARLY_15D_ADEQUATE'] = (df['EARLY_15D_PAYMENT_RATIO'] >= 0.5).astype(int)

    # 3. Initial pay vs deposit differential
    df['INITIAL_DEPOSIT_GAP'] = df['INITIAL_PAY'] - df['DEPOSIT']
    df['INITIAL_DEPOSIT_RATIO'] = df['INITIAL_PAY'] / (df['DEPOSIT'] + 1)

    return df
```

#### 8. Credit Check Features (`engineer_credit_check_features`)
These features encode information from the credit check process.

*   `CREDIT_CHECK_ENCODED`: Numerical encoding of `CREDIT_CHECK_DONE` (Pass=2, Pending=1, Fail/Not Done=0).
*   `NO_CREDIT_CHECK`: Binary flag (0 or 1) if a credit check was not 'Pass' or 'Pending'.

```
def engineer_credit_check_features(credit_df):
    """Features based on credit check results."""
    df = credit_df.copy()

    # Encode credit check results
    credit_check_map = {
        'Pass': 2,
        'Pending': 1,
        'Fail': 0,
        'Not Done': 0
    }
    df['CREDIT_CHECK_ENCODED'] = df['CREDIT_CHECK_DONE'].map(credit_check_map).fillna(0)

    # Flag for missing/insufficient credit check
    df['NO_CREDIT_CHECK'] = (~df['CREDIT_CHECK_DONE'].isin(['Pass', 'Pending'])).astype(int)

    return df
```

```
# ============================================
# APPLY ALL FEATURE ENGINEERING
# ============================================

def engineer_all_features(credit_df):
    """Apply all feature engineering functions."""
    print("Starting feature engineering...")

    df = credit_df.copy()

    # Store original columns to identify new features
    original_cols = set(df.columns)

    # Apply all feature groups
    df = engineer_target_variables(df)
    df = engineer_payment_behavior_features(df)
    df = engineer_temporal_features(df)
    df = engineer_financial_ratios(df)
    df = engineer_risk_scores(df)
    df = engineer_first_payment_features(df)
    df = engineer_credit_check_features(df)

    # Identify new features
    new_cols = set(df.columns) - original_cols
    print(f"\nCreated {len(new_cols)} new features")
    print("\nNew features:")
    for col in sorted(new_cols):
        print(f"  - {col}")

    return df

# Apply all feature engineering
credit_engineered = engineer_all_features(credit)
```
